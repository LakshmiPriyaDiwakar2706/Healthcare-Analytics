```python
# create a new column age from birthdate 

# but if we have death date anytime before now use that to calculate age 

def calculate_age(row):

    current_date = pd.Timestamp('now')
    birth_date = row['birthdate']
    death_date = row['deathdate']

    if death_date <= current_date:
        age = (death_date - birth_date).days // 365  #converting it to years

    else:
        age = (current_date - birth_date).days // 365

    return age

patients.insert(3,'age',patients.apply(calculate_age,axis=1)) # apply function with axis=1 passes one row at a time to the function. each row is passed as a pandas series, where each column's vlue can be accessed by the column name
```
